```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
with open('15.txt', 'r') as f:
    input_data = f.read().split('\n')
```

```python
len(input_data)
```

```python
di = dict()
```

```python
index = 0
while index < len(input_data):
    a, n, noize, cnt = map(float, input_data[index].split())
    index += 1
    a, n, cnt = int(a), int(n), int(cnt)
    if noize not in di:
        di[noize] = dict()
    len_mas = int(input_data[index])
    index += 1
    mas = []
    for i in range(len_mas):
        q, w = map(int, input_data[index].split())
        index += 1
        mas.append(np.array([q, w]))
    di[noize][a] = np.array(mas)
```

```python
di
```

```python
np.quantile(di[0.1][5][:, 1], 0.99)
```

```python
fig, ax = plt.subplots(figsize=(40,10))
width = 0.7
A = 2
N = 15
noize = 0.1

ax.bar(di[noize][A][:, 0], di[noize][A][:, 1], width, label=f'noize={noize}')
# plt.axhline(y=np.quantile(di[noize][A][:, 1], 0.999), xmin=0, xmax=4096, color='r')

ax.set_ylabel('times')
ax.set_xlabel('bit')
plt.xlim([-5, 256])
plt.title(f'   a = {A}, N = {N}')
# ax.set_xticks([i for i in range(256)])
ax.legend()
plt.savefig(f'./figs/a={A},N={N},noize={noize}.jpg')
```
